# **Creating a new environment:**


An *environment* contains all the installed packages needed for a particular script, model etc. 
You can select different environments to run different scripts. 
This can be done by selecting the kernel thing at the top right.
All my environments are stored in my UKCEH folder .venv.
To create a new environment, you need to use a computer terminal. It is recommended that Windows PowerShell is used as Command Prompt is a legacy terminal which doesn't work very well anymore. 

### **Code to put in the computer terminal to create a new environment**

*Example "start" directory (for me)*: **PS C:\Users\sopgra>**

1) Navigate into your new folder directory (in your PC, not in the cloud)

```python
PS C:\Users\sopgra> cd C:\Users\sopgra\UKCEH
```

2. If you want to create a new folder to store your **PYTHON** environments in, you can do it like this:

```python
PS C:\Users\sopgra\UKCEH> py -m venv .venv
```

3. You may need to permit yourself to execute certain commands, like for installing packages;

```python
PS C:\Users\sopgra\UKCEH> Set-ExecutionPolicy -ExecutionPolicy RemoteSigned -Scope CurrentUser
```

4. This next bit then tells the computer to refer to the version of python within your .venv environment, rather than the system version of python. Therefore, any packages that are installed whilst in this environment will only install the package into this environment, rather than globally

```python
PS C:\Users\sopgra\UKCEH> .venv\Scripts\Activate.ps1
```

5. You can then install your modules and libraries that you want in your environment. It is worth ensuring your pip installer is updated. This code can also be used to update any other pre-installed packages.

```python
(.venv) PS C:\Users\sopgra\UKCEH> py -m pip install --upgrade pip
(.venv) PS C:\Users\sopgra\UKCEH> pip install numpy pandas matplotlib pillow # etc. etc. (add any other packages you need for your project)
```
