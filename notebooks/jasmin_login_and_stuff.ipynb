# How to access JASMIN from the terminal

I use Git Bash becuase PowerShell is shit apparently

```python
# 1) Start the SSH Agent up
eval "$(ssh-agent -s)"
```

```python
# 2) Add my JASMIN key / identity 
ssh-add ~/.ssh/id_ecdsa_jasmin
```

```python
# 3) Check it loaded
ssh-add -1
```

```python
# 4) Connect to JASMIN login with agent forwarding
ssh -A sophiag@login.jasmin.ac.uk
# you will be asked for a pass key which is 12345 for me!
```

This is how to use GitBash to transfer files from the local folder to JASMIN (without having to drag and drop)
