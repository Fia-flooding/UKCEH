# Submitting a script to JASMIN (sbatch)

```python
# 1) start Git Bash and connect to JASMIN with SSH forwarding enabled:
ssh -A sophiag@login.jasmin.ac.uk
```

```python
# 2) Connect to a sci server once on the login node
ssh sci-ph-01
```

```python
# 3) Check which server you're on:
hostname

# expected output:
    # sci-ph-01
```

```python
# 4) Activate Python environment
# 4.1) load JASPY
module load jaspy

# 4.2) activate my virtual environemnt (venv)
source /home/users/sophiag/UKCEH/notebooks/ukceh_env/bin/activate
# edit this depending on which venv you use and where it is stored

# 4.3) check this
which python

# expected output:
    # /home/users/sophiag/UKCEH/notebooks/ukceh_env/bin/python

# 4.4) Check the available packages
python -c "import numpy,pandas,xarray,scipy; print('success')"
```

Part 2: Submit the job and monitor the job

```python
# template:
# sbatch my_job.sbatch

# example:
# sbatch UKCEH/slurm/trial_slurms/trial_blended_eof_slurm.sbatch

# example output:
# Submitted batch job 40188590
```

```python
# Check if the job is still running
squeue -j 40188590

# Get more detailed info:
scontrol show job 40188590

# See all your jobs
squeue -u $USER

# Check whether it finished
sacct -j 40188590
```

```python
# If the job failed, check:
cat ~/logs/errors_40188590.err
cat ~/logs/outputs_40188590.out
```

```python
# Cancel a job
scancel 40188590
```

How to check a job after re-opening Git Bash

```python
# 1) log back into JASMIN and a sci server (See above)

# 2) Check whether the job is still running:
squeue -j 4-01889590

# output
# if ST = R, the job is running
# if nothing appears, the job is finished
```
